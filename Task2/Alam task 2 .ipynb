```python
import pandas as pd
import hashlib
```

```python
df = pd.read_excel('mobile_customers.xlsx')
df.head()
```

```python
df_anonymized=df
df_anonymized.head()
```

```python
df.shape
```

```python
# Step 1: Remove unnecessary columns
df_anonymized = df.drop(columns=['Unnamed: 0', 'name', 'address', 'email', 'credit_card_number', 'credit_card_security_code'])
df_anonymized.shape
```

```python
# Step 2: Mask sensitive data
# Hash the customer_id
df_anonymized['customer_id'] = df_anonymized['customer_id'].apply(lambda x: hashlib.sha256(x.encode()).hexdigest())
df_anonymized.head()
```

```python
# Mask credit card provider (e.g., showing only first 4 characters)
df_anonymized['credit_card_provider'] = df_anonymized['credit_card_provider'].apply(lambda x: x[:4] + '****')
df_anonymized.head()
```

```python
# Mask credit card expiration date
df_anonymized['credit_card_expire'] = df_anonymized['credit_card_expire'].apply(lambda x: '****-**')
df_anonymized.head()
```

```python

# Step 1: Categorize 'age' into brackets
age_bins = [0, 18, 25, 35, 45, 55, 65, 100]  # Age ranges
age_labels = ['<18', '18-25', '26-35', '36-45', '46-55', '56-65', '65+']
df_anonymized['age_bracket'] = pd.cut(df['age'], bins=age_bins, labels=age_labels, right=False)
df_anonymized=  df.drop(columns=['age', 'salary'])
# Step 2: Categorize 'salary' into brackets
salary_bins = [0, 50000, 100000, 150000, 200000, 300000]  # salary ranges
salary_labels = ['0-50K', '50K-100K', '100K-150K', '150K-200K', '200K+']
df_anonymized['salary_bracket'] = pd.cut(df['salary'], bins=salary_bins, labels=salary_labels, right=False)
```
